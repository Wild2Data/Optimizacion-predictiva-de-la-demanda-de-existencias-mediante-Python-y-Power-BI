Este notebook presenta un análisis completo de ventas, inventario y estacionalidad
en un negocio retail de surf con múltiples sucursales en España, con el objetivo
de **anticipar quiebres de stock y apoyar la toma de decisiones de negocio**.


##  Contexto del negocio

Atlantic Flow Surf es una tienda especializada en productos y servicios de surf,
con presencia en tres ciudades costeras de España:

- San Sebastián
- Santander
- Fuerteventura

El negocio combina:
- Venta de productos de surf
- Clases de surf para distintos niveles
- Gestión de inventario multisucursal

Debido a la **alta estacionalidad** y a la variación de la demanda,
la empresa enfrenta problemas recurrentes de **quiebres de stock**,
afectando ventas y experiencia del cliente.


##  Problema de negocio

La empresa no cuenta con una forma clara de responder preguntas clave como:

- ¿Qué productos tienen mayor riesgo de quiebre de stock?
- ¿En qué sucursales y temporadas ocurre con más frecuencia?
- ¿Cómo influyen las ventas recientes en los quiebres?
- ¿Podemos anticipar estos problemas antes de que ocurran?

Actualmente, las decisiones se toman de forma reactiva,
lo que genera pérdidas de ventas y sobrecostes logísticos.


##  Objetivo del análisis

El objetivo de este proyecto es:

- Analizar el comportamiento histórico de ventas e inventario
- Identificar patrones de estacionalidad y rotación
- Construir un dataset preparado para modelos predictivos
- Anticipar quiebres de stock mediante análisis de datos

Este análisis busca **transformar datos en decisiones accionables**.


##  Fuente de datos

Los datos utilizados en este análisis provienen de un
**Data Warehouse diseñado en SQL Server**, con:

- Más de 3 años de información diaria
- Ventas, clases e inventario
- Datos simulados con lógica de negocio realista
- Quiebres de stock controlados (~12%)

El dataset final está construido a nivel:
**producto – sucursal – día**.


##  Preparación del dataset

Se utiliza una vista analítica unificada que integra:

- Ventas diarias
- Estado de inventario
- Variables temporales (mes, temporada)
- Variable objetivo: quiebre de stock (sí / no)

El resultado es un dataset limpio, consistente y listo
para análisis exploratorio y modelado predictivo.


##  Análisis exploratorio de datos

En esta sección se analizan:

- Distribución de ventas en el tiempo
- Diferencias entre sucursales
- Relación entre ventas e inventario
- Frecuencia de quiebres de stock
- Impacto de la estacionalidad

El objetivo es **entender el problema antes de modelar**.


##  Insights clave

Algunos hallazgos relevantes:

- Los quiebres de stock se concentran en temporadas de alta demanda
- No todos los productos se venden diariamente (rotación realista)
- Las sucursales turísticas presentan mayor presión sobre inventario
- Existe relación entre ventas recientes y quiebres posteriores

Estos insights permiten enfocar mejor la predicción y la planificación.


##  Próximos pasos

A partir de este análisis, los siguientes pasos son:

- Feature engineering (lags, medias móviles)
- Entrenamiento de modelos predictivos
- Evaluación del impacto en decisiones de inventario
- Visualización de resultados en dashboards

El objetivo final es **anticiparse al problema, no reaccionar a él**.
